```python
beer_data = pd.DataFrame()
beer_features = ['id', 'name', 'nameDisplay', 'abv', 'styleId', 'isOrganic', 'isRetired', 'labels', 'status', 'statusDisplay', 'createDate', 'updateDate', 'style']
for feature in beer_features:
    beer_data[feature]=''

count=0
for p in tqdm(range(1,24)):
    data = ap.get_data(page=p)
    for i in range(50):
        for feature in beer_features:
            try:
                beer_data.set_value(count,feature,data['data'][i][feature])

            except Exception as e: 
                print(f"name:{data['data'][i]['name']}, page:{p}, item:{i}, feature:{feature}, error:{e}")
        count+=1
```
